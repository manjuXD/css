```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('NationalElectionIndia.xlsx')
df['Win'] = 0
```

# Computing the Winners

```python
wonlist=list(df.groupby(['year','pc_name'])['totvotpoll'].idxmax())
df.loc[df.index.isin(wonlist),'Win']= 1
```

```python
Years = list(pd.unique(df.year)) 
Years
```

```python
winnerdf_female = df.copy()
```

```python
national_party_df = pd.read_excel('List of Political Parties Participated 2014.xlsx')
national_party_df = national_party_df.fillna(0)
d = {}
for x in Years:
    temp = list(national_party_df[x])
    l = []
    for j in temp:
        if j != 0:
            l.append(j.split(',')[0])
    d[x] = l
```

```python
d  # Dictonary containing National Parties in a particular year
```

# Computing the party of the candidate

```python
for year in Years:
    temp=d[year]
    winnerdf_female.loc[(winnerdf_female.partyabbre.isin(temp)) & (winnerdf_female.year == year) , 'party_type'] = 'N'

winnerdf_female.loc[winnerdf_female.partyabbre=='IND', 'party_type'] = 'I'
winnerdf_female["party_type"].fillna("S", inplace = True)
```

# Visualization of Winning women parties 

```python
final_dic_year = {}
for x in Years:
    Indep=[]
    national=[]
    state=[]

    df_participation_female = winnerdf_female[(winnerdf_female['cand_sex'] == 'F') & (winnerdf_female['year']==x)]
    df_won_female =  winnerdf_female[(winnerdf_female['cand_sex'] == 'F') & (winnerdf_female['Win']==1) & (winnerdf_female['year']==x)]
    d1_f_won = dict(df_won_female['party_type'].value_counts())
    d2_f_part = dict(df_participation_female['party_type'].value_counts())
    if 'I' not in d1_f_won.keys():
        d1_f_won['I'] = 0
    #print (d1_f_won,d2_f_part)

    national.append((d1_f_won['N']/d2_f_part['N'])*100)   #National
    state.append((d1_f_won['S']/d2_f_part['S'])*100)   #state
    Indep.append((d1_f_won['I']/d2_f_part['I'])*100)   #Independent

    final_dic_year[x] = national+state+Indep

#print ("Percentage of National,State,Independent Winning women")
#final_dic_year
```

```python
Indep=[]
national=[]
state=[]
for x in final_dic_year.values():
    national.append(x[0])
    state.append(x[1])
    Indep.append(x[2])

plt.rcParams["figure.figsize"] = [16,9]

party_type=Years

xpos = np.arange(len(Years))

plt.bar(xpos+0.010,national, width=0.25, label="National")
plt.bar(xpos+0.26,state, width=0.25,label="State")
plt.bar(xpos+0.51,Indep, width=0.25,label="Independent")

plt.xticks(xpos,Years)
plt.ylabel("Number of women winners")
plt.title('Winning women representation from political party')
plt.xlabel('Election Year')
plt.legend()
plt.show()
```

# Performance of Women in Election Seats

```python
x = winnerdf_female.loc[winnerdf_female['cand_sex'] == 'F'].copy()
x['cand_sex'] = 1                                               #Female candidates
z = x.groupby(['year']).sum()                                   #Calculating number of females           
```

```python
z = z.drop(['pc_no','totvotpoll','electors'], axis=1)                 # Columns not required
z['seats'] = [542,542,514,529,407,543,543,543,543,543,543]            # Total Number of seats per year
z['Productivity_gap']=round((z['Win']/z['cand_sex'])*100,2)
z["years"] = list(z.index)
z
```

```python
df = winnerdf_female.copy()

df_cross=pd.crosstab(df["year"],df["cand_sex"],margins=True)

df_cross['Participation_gap']=round((df_cross['F']/(df_cross['M']+df_cross['O']))*100,2)

#df_cross=df_cross.drop(['All'])

#df_cross['norm']=((df_cross['p_gap']-df_cross['p_gap'].min())/(df_cross['p_gap'].max()-df_cross['p_gap'].min())).round(decimals=3)

#df_cross['cent_country']=((df_cross['F']/df_cross['All'])*100).round(decimals=2)

z['Participation_gap'] = df_cross['Participation_gap']
z
```

```python
z.index
```

```python
df2 = z.groupby(['years'])[['Participation_gap','Productivity_gap']].sum().plot(kind='bar', stacked=False, figsize=(20,20), fontsize = 28, rot = 0)


df2.set_ylabel('Percentage %', fontsize = 28)
df2.set_xlabel('Election Years', fontsize = 28)
df2.set_title('Representation of Women in Indian Elections', fontsize = 35)
df2.legend(bbox_to_anchor=(1, 1), fontsize = 28)

for p in df2.patches:
    df2.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize = 28)

plt.savefig('css2.png', dpi=1000)
```
